### Paso 3. Asígnalo a una variable llamada usuarios.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
usuarios = pd.read_csv (url, sep = '|')
usuarios.head()
```

```python
usuarios.columns
```

### Paso 4. Descubre cuál es la edad media por ocupación

```python
usuarios.groupby(['occupation'])[['age']].mean().head()
```

### Paso 5. Descubra la proporción de Hombres por ocupación y ordénela de mayor a menor

#### Forma 1

```python
usuarios1 = usuarios.groupby(['occupation', 'gender'])[['gender']].value_counts()
#print(usuarios1)
#usuarios1.groupby(['occupation'])['gender'].agg()
usuarios1 = pd.DataFrame(usuarios1)
k=usuarios1.unstack()
k['poporcion_m'] =k['count']['M']/(k['count']['F']+k['count']['M'])
k
```

### Paso 6. Para cada ocupación, calcule las edades mínima y máxima
